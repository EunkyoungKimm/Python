## "연습" 영화랭킹의 5개 페이지 파싱

- 현재 랭킹 페이지: https://movie.naver.com/movie/sdb/rank/rmovie.naver
- 이전 날짜 랭킹 페이지: https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cnt&tg=0&date=20221120
   - `URL?date=yyyymmdd`형식

```python
from bs4 import BeautifulSoup
import requests
```

```python
! pip install bs4 html5lib lxml chardet
```

```python
dates = ['20221121', '20221120', '20221119', '20221118', '20221117']

results = []


for d in dates:
    # Soup crawling
    response = requests.get('https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cnt&tg=0&date=' + d)
    soup = BeautifulSoup(response.text, 'lxml')
    # tit3 클래스
    tags = soup.select('div[class=tit3]')
    print('<'+d+' 영화랭킹>')
    
    for tag in tags:
        # results append
        results.append(tag.text.strip())
        
    for k,i in enumerate(results):
        print(k+1,i)
```

## 영화랭킹 - 함수 같이 정리해보자!

```python
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://movie.naver.com/movie/sdb/rank/rmovie.naver')
response.status_code
```

```python
soup = BeautifulSoup(response.text)
```

```python
tags = soup.find_all('div', attrs={'class':'tit3'})
results = []
for idx,tag in enumerate(tags): # 리스트(tags)를 enumerate하면 index값을 얻을 수 있다! index 얻을때 enumerate많이씀
    # print(tag.text.strip())
    results.append( {idx: tag.text.strip()} ) # 번호 : 제목 (딕셔너리가 하나씩 들어감!)
```

```python
import csv
with open('data/movie_ranking.csv', 'w', encoding='utf-8') as csvfile: # utf-8하면 VSC에서는 O, 엑셀은 X / euc-kr하면 VSC에서는 X, 엑셀은 O
    spamwriter = csv.writer(csvfile) # delimiter 콤마가 기본값
    spamwriter.writerow(['번호','제목']) # header가 써짐 - 번호, 제목
    # for
    for idx,row in enumerate(results):
        spamwriter.writerow([idx, row[idx]]) # 번호, 제목 / 딕셔너리니까 row[idx]
```
