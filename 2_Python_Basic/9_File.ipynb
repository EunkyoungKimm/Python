## open()

파일은 open() 명령에서 `<MODE>` 값에 따라 읽고 쓴다. [Python builtin functions](https://docs.python.org/3/library/functions.html)
    
>open(file, mode='rb', buffering=-1, encoding=None, errors=None, newline=None, closefd=True, opener=None)
  - mode는 파일에 대한 액션과 파일 형식을 지정한다: rb, wb, rt, wt
     - 액션 : w(기록), a(추가), r(읽기:기본값)
     - 타입 : t(텍스트:기본값), b(2진 파일)
open 으로 열린 객체는 `.close()` 로 닫아 주어야 한다.

file:
- Text file: 문자코드(ascii code (를 기반으로 한)), ipynb도 text파일임
- Binary file: byte streams(바이트 단위로 저장되어 있음, 이미지 등)

`readline()` 함수 이용하기

파일이 클 경우, 라인 단위로 파일을 읽는 것이 효율적
- 읽기 함수: f.read([size]) 문자 수 만큼 / f.readline([size]) 몇 줄.. 라인 수 만큼 / f.readlines([size]) 전체 라인을 리스트로 반환

위키백가 화이트스페이스 보기

```python
f = open("test.txt", "rt", encoding="utf8")
while True:
    line = f.readline() # --> EOL \n \r\n EOF 힌줄씩 읽히는거..? 라인단위로 보고싶으면 이걸로!
    if not line: # 공백이 오지 않을때까지 읽는거임
        break
    line = line.strip() # 공백 문자 제거(strip 왼쪽, 오른쪽 공백 지움..)
    print(line)

f.close()
```

```python
# 강사님

#파일 읽기

#readines()로 모두 읽기
    
# 리스트 크기 만큼 반복
    # 문자열을 ' ' 분리 -> split
    # 결과를 len() -> ++sum

with open("Q-sequence실습.ipynb", "r", encoding='utf8') as f:
    line = f.readlines()
    count = 0
    for i in range(len(line)):
        text = line[i]
        # print(text)
        word = []
        for k in range(len(text)):
            if text[k] != '':
                word.append(text[k]) 
        count += len(word) # 들여쓰기 중요
    print(count)
```

## binary mode

바이너리 파일은 바이너리 모드로 작성해서 사용해야 한다. 

```python
"""
바이너리 파일 생성
"""
print("====== binary file =====")
f = open("test.bin", "wb")

for i in range(0, 256):
    data = i.to_bytes(1, 'little')
    f.write(data)
f.close()
```

```python
a = 1
a.to_bytes(1, 'little') # little endian
```

- little endian big endian (비트도..순서o 순서에 따라 크기, 값 !=) / os 표준을 만들때 ..little endian이 디폴트값

```python
# with
print("====== with binary file =====")
with open("test.bin", "wb") as f: # wb -> 바이너리 형태로 저장됨

    for i in range(0, 256):
        data = i.to_bytes(1, 'little')
        f.write(data)
```

```python
with open("test.bin", "rb") as f:
    print("{0}".format(f.read()).encode()) # 문자열화가 된걸 문자코드로 변환해줌
#     print(f'{f.read()}'.encode()) # a byte string
```
