```python

import requests
from bs4 import BeautifulSoup

# 대분류
print('100=정치, 101=경제, 102=사회, 103=생활문화, 104=세계, 105=IT/과학')
main_category = input('코드번호 입력:')

# 중분류
print('정치 >> 264=청와대, 265=국회/정당, 268=북한, 266=행정, 267=국방/외교, 269=정치일반\n'
      '경제 >> 258=증권, 261=산업/재계, 771=중기/벤처, 260=부동산, 262=글로벌 경제, 310=생활경제, 263=경제 일반')
middle_category = input('코드번호 입력:')

url = f'https://news.naver.com/main/list.nhn?mode=LS2D&mid=shm&sid1={main_category}&sid2={middle_category}'
    
```

```python
url
```

```python
 
r = requests.get(url, headers={'user-agent':'chrome'})
soup = BeautifulSoup(r.text)

articles1 = soup.select('ul[class=type06_headline]')
articles2 = soup.select('ul[class=type06]')
articles = articles1 + articles2

article_lst = []
for article in articles:
    # title
    title = article.select('dt:nth-child(2) > a')[0].text
    # text
    article_lst.append(title)
```
